```python
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(15,10)})
```

```python
df = pd.read_csv('../EVERYTHING/voron.csv').sort_values('unique users (число пользователей, прошедших этап)', ascending = False)
```

```python
df.head(2)
```

```python
sns.barplot(data = df, x = 'unique users (число пользователей, прошедших этап)', y = 'event_name', palette = "dark")
```

```python
df['win_uniq'] = df['unique users (число пользователей, прошедших этап)'].shift(+1)
#Сдвигаю столбец с данными, таким образом чтобы посчитать ретеншн
```

```python
df['retention,%'] = round(df['unique users (число пользователей, прошедших этап)'] /df.win_uniq   * 100, 2)
#Ретеншн, %
```
